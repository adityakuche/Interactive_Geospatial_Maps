```python
import pandas as pd
import numpy as np
  
dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_forests_in_India')
len(dfs)
```

```python
df=dfs[0]
df
```

```python
df= df.drop(columns=['Image','Comments'],axis=1)
```

```python
dfa= df.dropna(axis=0,subset=['Area'])
dfa
```

```python
#dfa.shape #(21, 3)
#dfa.to_csv('dfa.csv') #save the DataFrame into the csv
# calcualting the lat, long from the name columns
# then plot the national park using folium library
```

```python
from geopandas. tools import geocode 
```

```python
#Calculating lat/long according to the name of national parks in the row
for index, row in dfa.iterrows():
    print (row['Name'])
    
    info= geocode(str(row['Name']), provider='arcgis',timeout=10)
    #data_cleaned.loc[int(index), 'Address']= info['address'].loc[0]
    dfa.loc[int(index), 'lon']=info['geometry'].loc[0].x
    dfa.loc[int(index), 'lat']=info['geometry'].loc[0].y
```

```python
dfa
```

```python
dfa.to_csv('List of forests in India.csv')
```

```python
import folium
from folium.plugins import HeatMap
```

```python
locations = list(zip(dfa['lat'], dfa['lon']))
map_1= folium.Map([28.59, 78.96], zoom_start=5)
HeatMap(locations).add_to(map_1)
map_1
```

```python
popup= [names for names in dfa['Area']]


from folium.plugins import MarkerCluster
m = folium.Map(location=[28.59, 78.96], zoom_start=5)
marker_cluster =  MarkerCluster(
    locations=locations, popups=popup,
    overlay=True,
    #control=True,
    #icon_create_function=icon_create_function
)
marker_cluster.add_to(m)
m
```
